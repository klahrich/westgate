```python
import pandas as pd
```

```python
month_yyyymm = '2023-04'
month_str = 'apr2023'
```

```python
basefolder = 'C:/Users/karim/Documents/westgate/data/training/'
```

```python
data2023_df = pd.read_csv(basefolder + 'janToNov2023LoansAccepted.csv')
```

```python
month_df = data2023_df.loc[data2023_df['request_date'].str[:7]==month_yyyymm]
```

```python
len(month_df)
```

```python
month_df.drop_duplicates(inplace=True)
```

```python
len(month_df)
```

```python
month_df.head()
```

```python
month_df.set_index('LoanId', inplace=True)
```

```python
loan_outcomes_df = pd.read_csv(basefolder + 'loan_outcomes_' + month_str + '.csv')
```

```python
loan_outcomes_df['profit'] = loan_outcomes_df['Paid Amount'] - loan_outcomes_df['Principal Amount']
```

```python
loan_outcomes_df.set_index('Loan Id', inplace=True)
```

```python
loan_outcomes_df.head()
```

```python
merged_df = pd.merge(month_df, loan_outcomes_df, 
                     left_index=True, right_index=True, 
                     how='inner',
                     validate='one_to_one')
```

```python
len(merged_df)
```

```python
merged_df.reset_index(inplace=True)

merged_df.rename(columns={
    'Released Date': 'release_date',
    'Principal Amount': 'principal',
    'Paid Amount': 'total_paid',
    'index': 'Id',
    'Loan Status Name': 'status'
    },
    inplace=True
)
```

```python
merged_df.head()
```

```python
merged_df['status'].value_counts()
```

```python
merged_df.head()
```

```python
merged_df[['Id', 'principal', 'total_paid', 'profit', 'status']].head()
```

```python
merged_df.to_csv(basefolder + 'loan_outcomes_' + month_str + '.csv', index=False)
```
